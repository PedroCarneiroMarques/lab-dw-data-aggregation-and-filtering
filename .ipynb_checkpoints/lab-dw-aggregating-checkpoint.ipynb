This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)

missing_values = df.isnull().sum() # 0
missing_values
df = df.drop(df.columns[0], axis=1).rename(columns={df.columns[0]: "unique_id"}) # renamed 1st column
df.columns = df.columns.str.lower().str.replace(' ', '_') # all non capitol letters
df.columns 
mean_income = round(df['income'].mean()) 
df['income'] = df['income'].replace(0, mean_income) # replaced "0" values with the mean rounded to 0 decimal places 
df = df.drop("vehicle_type", axis=1) # dropped vehicle_type because of too many NaN values
pd.set_option('display.max_columns', None)

df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

df
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]
result = df.groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
result = result.reset_index()
result

# The table shows 6 entries: 2 of each of the 3 policy_tipes 
# and 3 of each gender gender that surpassed the >1000 total_claim_amount mark  
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
customer_state = df['customer'].groupby(df['state']).count().reset_index()
plus_five_hundred = customer_state[customer_state['customer'] > 500]
print(plus_five_hundred)

# there are no states with >500 customers
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
ed_gender_level = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).reset_index()
ed_gender_level

# the main conclusion is that females tend to have a greater median life_time_values across all education levels
# therefore should be the main focus for the target population 
```
